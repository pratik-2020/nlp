```python
import pandas as pd
import numpy as np
def rmv(st):
    ar = ['.',',',';',':','!','@','$','#','%','^','&','*','(',')','-','_','=','+','/','<','>','?','"','{','}','[',']',"'"]
    ans = ""
    for i in st:
        if i in ar:
            continue
        else:
            ans = ans + i
    return ans.lower()
dst = pd.read_csv("C:/Users/cw/Desktop/nis/dialogs_expanded.csv")
vb = []
st = set([])
dct = {}
dct['unk'] = 0
for i in range(20000):
    ar = dst['question'][i].split(' ')
    for j in ar:
        f = rmv(j)
        if not f in st:
            dct[f] = 0
        st.add(f)
        dct[f] = dct[f] + 1
for i in st:
    if dct[i] >= 2:
        vb.append(i)
    else:
        dct['unk'] = dct['unk'] + 1
# vb.append('unk')
# print(vb)
mtr = np.zeros((len(vb)+3, len(vb)+3))
for i in range(20000):
    ar = dst['question'][i].split(' ')
    for j in range(len(ar)):
        if j == 0:
            f = rmv(ar[j])
            if f in vb:
                mtr[0][vb.index(f)] = mtr[0][vb.index(f)] + 1
            else:
                mtr[0][2] = mtr[0][2] + 1
        else:
            f = rmv(ar[j])
            if f in vb:
                if rmv(ar[j-1]) in vb:
                    mtr[vb.index(rmv(ar[j-1]))][vb.index(f)] = mtr[vb.index(rmv(ar[j-1]))][vb.index(f)] + 1
                else:
                    mtr[2][vb.index(f)] = mtr[2][vb.index(f)] + 1
            else:
                if rmv(ar[j-1]) in vb:
                    mtr[vb.index(rmv(ar[j-1]))][2] = mtr[vb.index(rmv(ar[j-1]))][2] + 1
                else:
                    mtr[2][2] = mtr[2][2] + 1
# print(mtr)
for i in range(len(mtr)):
    for j in range(len(mtr[0])):
        if i <= 2:
            mtr[i][j] = np.log(mtr[i][j]/20000)
        else:
            mtr[i][j] = np.log((mtr[i][j] + 1)/(dct[vb[i-3]] + len(vb)))
# print(mtr)
def fun(s,pm):
    ar = s.split(' ')
    pr = 0
    for i in range(len(ar)):
        f = rmv(ar[i])
        if(i == 0):
            if not f in vb:
                pr = pr + pm[0][2]
            else:
                pr = pr + pm[0][vb.index(f)]
        elif i > 0:
            if not f in vb:
                if not rmv(ar[i-1]) in vb:
                    pr = pr + pm[2][2]
                else:
                    pr = pr + pm[vb.index(rmv(ar[i-1]))][2]
            else:
                if not rmv(ar[i-1]) in vb:
                    pr = pr + pm[2][vb.index(f)]
                else:
                    pr = pr + pm[vb.index(rmv(ar[i-1]))][vb.index(f)]
    return pr
def gen(s,pm,pb):
    ar = s.split(' ')
    f = 0
    if not rmv(ar[len(ar) -1]) in vb:
        f = 2
    else:
        f = vb.index(rmv(ar[len(ar)-1]))
    mx = pb + pm[f][3]
    ans = vb[3]
    for j in range(len(vb)):
        if(max(pb+pm[f][j], mx)) == pb+pm[f][j]:
            ans = vb[j]
            mx = pb+pm[f][j]
    return ans
s = input("Enter a string :- ")
g = fun(s,mtr)
print(gen(s,mtr,g))
```
